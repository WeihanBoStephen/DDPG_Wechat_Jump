# This code is for collect real-world training data for 'UARM' robot arm.

```python
import opencv_jump as opencv
from game_environment import Jump_Env as wechat_env
from robot_arm import uarm
import cv2
import time
import pymysql
```

```python
def get_time():
    return str(int(time.time()))
def save_nalu_img(img_state):
    state_dir = "./uarm_reality_data/"+get_time() + ".jpg"
    cv2.imwrite(state_dir, img_state)
    return state_dir
```

```python
def insert(action,img):
    action = str(action)
    img = str(img)
    db = pymysql.connect("localhost", "root", "root", "test")
    cursor = db.cursor()
    insert = "INSERT INTO real_word_train_data(img,action) VALUES('%s','%s')"%(img,action)
    result = cursor.execute(insert)
    db.close()
    print(insert)
    return result
def collect_data(action):
    img = cv2.imread("autojump.jpg")
    nalu_img_dir = save_nalu_img(img)
    insert(action,nalu_img_dir)
```

```python
arm = uarm()
number_templet = [cv2.imread('templet/{}.jpg'.format(i)) for i in range(10)]
restart_templet = cv2.imread('templet/again.jpg')
env = wechat_env(number_templet=number_templet, restart_templet=restart_templet)
while 1:
    wechat_score = 0
    jump_step = 0
    obs0 = env.reset()
    while 1:
        action = opencv.get_action("autojump.jpg")
        print("action:%f"%action)
        obs1, reward, done, score = env.step_robot(action,arm)
        if done:
            break
        else:
            collect_data(action)
            wechat_score = score
            jump_step += 1
            obs0 =obs1
```
