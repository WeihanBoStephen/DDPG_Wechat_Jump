# This code is for loss function and reward visualisation during model training

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import spline
```

```python
def smooth(data,weight=0.85):
    scalar = data['Value'].values
    last = scalar[0]
    smoothed = []
    for point in scalar:
        smoothed_val = last * weight + (1 - weight) * point
        smoothed.append(smoothed_val)
        last = smoothed_val
    result = pd.DataFrame({'Step':data['Step'].values,'Value':smoothed})
    return result
```

```python
def visulisation(title,data,dir_name,label,weight,x_name,y_name):
    data_new = smooth(data,weight=weight)
    x = data['Step'].values
    y = data['Value'].values
    x_new = data_new['Step'].values
    y_new = data_new['Value'].values
    plt.figure(figsize=(8,8))
    #plt.style.use(style)
    plt.title(title,fontsize=18)
    plt.plot(x, y, '#FFFF00', label=label)
    plt.plot(x_new, y_new, '#FF9900', label='After Smoothing=%.2f'%(weight),linewidth=2)
    plt.grid(True)
    plt.xlabel(x_name,fontsize=16)
    plt.ylabel(y_name,fontsize=16)
    plt.legend(loc="upper right")
    plt.savefig(dir_name)#"./Loss_Function_File/actor_loss_0_5759.jpg"
    plt.show()
```
